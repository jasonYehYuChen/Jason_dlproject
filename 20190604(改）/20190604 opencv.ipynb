# 1. 簡介

## 人體中的白血球種類可分為好幾種，本專案針對其中的四種白血球細胞(多核球、單核球各2種)進行分類：

### A) 多核球
#### 多核球的細胞質中含有若干微小囊性顆粒，顆粒中則貯有多種酶。按照顆粒的染色性質可分為：
* 嗜中性球：佔白血球總數 70 ～ 80 %，當有發炎反應時是第一個主動來參與的細胞。
* 嗜酸性球：可協助身體對抗引起過敏的過敏原，亦具有吞噬能力。

### B) 單核球
* 淋巴球：通常存在於淋巴系統中。血液中含有三種類型的淋巴球： B 細胞、 T 細胞和自然殺手細胞。 B 細胞為可以產生約束病原體的抗體。 CD4+ T 細胞配合免疫應答。 CD8+ T細胞（細胞毒素）和自然殺手細胞可以殺滅被病毒感染的體細胞。
* 單核球：在血液中是一個小小的單核細胞，然而一旦離開血管後，會轉變成為巨噬細胞，可隨時吞噬大量細菌。單核球擁有嗜中性球的「清理」功能，但是它比嗜中性球活得更長，因為它們擔負有其他的責任。單核球以及巨噬細胞向 T 細胞傳遞病原體的抗原片斷使得病原體能夠再次被識別和殺滅或加速身體的免疫應答。

## 6-1. 初始準備

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### 匯入圖片的套件

```python
import os     #可以匯入一個資料夾中的所有圖片
import cv2    #opencv 圖片數據處理工具
import scipy
from tqdm import tqdm
```

### 神經網絡的套件

```python
%env KERAS_BACKEND=tensorflow
from keras.models import Sequential
from keras.layers import Dense, Conv2D, BatchNormalization, Dropout, MaxPool2D, Input, Softmax, Activation, Flatten
from keras.models import Model
from keras import optimizers
from keras.layers import concatenate, AveragePooling2D
from keras.callbacks import ModelCheckpoint
from keras.utils.vis_utils import plot_model
from keras.optimizers import SGD, Adam
```

## 6-2. 匯入圖片數據

### 運用OpenCV，將白血球數據從資料夾中匯入 ，並建立他的代號。
* NEUTROPHIL(嗜中性球) = 1
* EOSINOPHIL(嗜酸性球) = 2
* MONOCYTE(單核球) = 3
* LYMPHOCYTE(淋巴球) = 4

```python
# 建立一個函數來匯入圖檔
def get_data(folder):
    # Load the data and labels from the given folder.
    X = []
    y = []
    for wbc_type in os.listdir(folder):
        if not wbc_type.startswith('.'):
            if wbc_type in ['NEUTROPHIL']:   # 嗜中性球
                label = 1
            elif wbc_type in ['EOSINOPHIL']:   # 嗜酸性球
                label = 2
            elif wbc_type in ['MONOCYTE']:   # 單核球
                label = 3  
            elif wbc_type in ['LYMPHOCYTE']:   # 淋巴球
                label = 4 
            else:
                label = 5   # 其他
            for image_filename in tqdm(os.listdir(folder + wbc_type)):
                img_file = cv2.imread(folder + wbc_type + '/' + image_filename)
                if img_file is not None:
                    img_file = scipy.misc.imresize(arr=img_file, size=(120, 160, 3))
                    img_arr = np.asarray(img_file)
                    X.append(img_arr)
                    y.append(label)
    X = np.asarray(X)
    y = np.asarray(y)
    return X,y
```

### 用上面建立的函數分別從資料夾中讀入Train 跟 Test 資料

```python
X_train, y_train = get_data('blood-cells/dataset2-master/images/TRAIN/')
X_test, y_test = get_data('blood-cells/dataset2-master/images/TEST/')
```

```python
# 儲存資料庫
np.save("X_train0_120_160.npy", X_train)
np.save("y_train0_120_160.npy", y_train)
np.save("X_test0_120_160.npy", X_test)
np.save("y_test0_120_160.npy", y_test)
```

### 6-2-2. 另一種匯入資料的方式

```python
X_train= np.load("X_train0_120_160.npy")
X_test= np.load("X_test0_120_160.npy")
y_train= np.load("y_train0_120_160.npy")
y_test= np.load("y_test0_120_160.npy")
```

### 將 label 數據轉換成 One Hot Encoding

```python
from keras.utils.np_utils import to_categorical
y_trainHot = to_categorical(y_train, num_classes = 5)
y_testHot = to_categorical(y_test, num_classes = 5)
```

### 檢查一下我們數據的格式及大小

```python
print("The shape of the X_training data: ", X_train.shape)
print("The shape of the y_training data: ", y_trainHot.shape)
print("The shape of the X_testing data: ", X_test.shape)
print("The shape of the y_testing data: ", y_testHot.shape)
```

## 6-3. 建立 CNN 模型(A)

### 利用 Functional API 建立模型(A)

```python
f1 = Conv2D(32, (3,3), padding="same",activation="relu")
f2 = MaxPool2D(pool_size=(2, 2), padding="same")
f3 = Conv2D(64, (3,3), padding="same",activation="relu")
f4 = MaxPool2D(pool_size=(2, 2), padding="same")
f5 = Dense(64,activation="relu")
f6 = Dropout(0.7)
f7 = Dense(5,activation="softmax")
```

```python
inp = Input(shape=(120,160,3))
k = BatchNormalization()(inp)  #normalization data
k1 = f1(k)
k2 = f2(k1)
k3 = f3(k2)
k4 = f4(k3)
k5 = Flatten()(k4)
k6 = f5(k5)
k7 = f6(k6)
y = f7(k7)
model1 = Model(inp, y)
model1.summary()
```

### 模型示意圖

![CNN model A](CNA.png)

## 6-3-1. 訓練網絡

```python
model1.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])
filepath = "CNNmodel_120_160.h5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]
model1_out = model1.fit(X_train,y_trainHot,epochs = 20, batch_size = 32,
                      validation_data = (X_test,y_testHot),
                      callbacks = callbacks_list,
                      verbose = 1)
```

### 儲存權重 

```python
from keras.models import load_model
model1.save('weight1_120_160.hdf5')
```

## 6-3-2. 對正確率作圖

```python
plt.plot(model1_out.history["acc"])
plt.plot(model1_out.history["val_acc"])
plt.title("model1 training accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(["training data accuracy", "testing data accuracy"], loc = "best")
plt.show()
```

## 6-3-3. 使用訓練好的權重

## 6-4. 建立 CNN 模型(B)

```python
f1 = Conv2D(64, (3,3), padding="same",activation="relu")
f2 = MaxPool2D(pool_size=(2, 2), padding="same")
f3 = Conv2D(128, (3,3), padding="same",activation="relu")
f4 = MaxPool2D(pool_size=(2, 2), padding="same")
f5 = Dense(128,activation="relu")
f6 = Dropout(0.7)
f7 = Dense(5,activation="softmax")
```
