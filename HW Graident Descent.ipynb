# 建立黑體輻射的公式

## 假設所有常數都是1 所以頻率與強度的關係為Ｉ＝ exp(v^3/e^v -1)


```python
def BK(v):
    
    return v**3/(np.exp(v)-1)
```

# Grandient Descent

```python
v=np.linspace(0.1,12,100)
HBK = BK(v)
plt.plot(v, HBK)
plt.ylabel("Intensity")
plt.xlabel("Wavelength")
plt.xlim(0, 12)
plt.ylim(0, 2)

Vst = input("起始點 {0.1~12}")
Vst = float(Vst)
VR = Vst+0.01
VL = Vst-0.01
EBKR = BK(VR)
EBKL = BK(VL)

err = 1
for i in range(0,20000):
    
    
    slope = (EBKR-EBKL)/(VR-VL)
    
    # learning rate
    η = -0.01
    Vmove = Vst-η*slope
    EBKmove = BK(Vmove)
    plt.scatter(Vmove, EBKmove)
    
    # 下一次的移動
    Vst = Vmove   
    EBKst = EBKmove
    VR = Vst+0.01
    VL = Vst-0.01
    EBKR = BK(VR)
    EBKL = BK(VL)
    
    
print(f"最大值{Vst}")

```
